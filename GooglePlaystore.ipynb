#3.1) How many free apps are there in ART_AND_DESIGN?

```python
c = df[(df['Category'] =='ART_AND_DESIGN') & (df['Type'] == 'Free')].shape[0]
print(f"There are {c} applications in ART_AND_DESIGN Category.")
```

##3.2) How many apps are there in ART_AND_DESIGN with rating more than 4.5?

```python
c1 = df[(df['Category'] == 'ART_AND_DESIGN') & (df['Rating'] > 4.5)].shape[0]
print(f"There are {c1} Applications in ART_AND_DESIGN Category with Rating above 4.5 ")
```

##3.3) How many apps are there in FAMILY with rating > 4.5 and Free?

```python
c2 = df[(df['Category'] == 'FAMILY') & (df['Rating'] > 4.5) & (df['Type'] == 'Free')].shape[0]
print(f"There are {c2} Applications in FAMILY category with Rating more than 4.5 are Free Type.")
```

##3.4) List all the free apps with rating more than 4.5 and category is FAMILY?

```python
l =[]
for i in df[(df['Category'] == 'FAMILY') & (df['Rating'] > 4.5) & (df['Type'] == 'Free')]['App']:
  l.append(i)
for i in l:
  print(i)
```

##3.5) Name the apps that are in ENTERTAINMENT with rating more than 4.5 in descending order WRT Ratings?


```python
 print(df[(df['Category'] == 'ENTERTAINMENT') & (df['Rating'] > 4.5)].sort_values('Rating', ascending=False)['App'])
```

##3.6) Which category is having maximum average rating Descending order?

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending= False).shape[0]
```

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending= False)
```

```python
df.groupby('Category')['Rating'].mean().sort_values(ascending= False).plot(kind = 'line')
```

##3.7) How many paid apps are there in each category in Descending Order?

```python
df[df['Type'] == 'Paid'].groupby('Category')['App'].count().sort_values(ascending = False)
```

```python
df[df['Type'] == 'Paid'].groupby('Category')['App'].count().sort_values(ascending = False).plot(kind = 'bar')
```

##3.8)How many Free apps are there in each category in Descending Order?

```python
df[df['Type'] == 'Free'].groupby('Category')['App'].count().sort_values(ascending = False)
```

```python
df[df['Type'] == 'Free'].groupby('Category')['App'].count().sort_values(ascending = False).plot(kind = 'bar')
```

##3.9) Name the Apps that are in ART_AND_DESIGN in descending order WRT Reviews.

```python
print(df[(df['Category'] =='ART_AND_DESIGN')].sort_values('Reviews', ascending = False)['App'])
```
